```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import cleaned IFL, PFL, and URBdata from github
url1 = 'https://raw.githubusercontent.com/adaire2/Project-2/main/DATA/clean_IFL.csv'
ifl = pd.read_csv(url1)

url2 = 'https://raw.githubusercontent.com/adaire2/Project-2/main/DATA/clean_PFL.csv'
pfl = pd.read_csv(url2)

url3 = 'https://raw.githubusercontent.com/adaire2/Project-2/main/DATA/clean_URB.csv'
urb = pd.read_csv(url3)
```

```python
ifl.head()
```

```python
ifl['country'].describe()
```

```python
# List of selected countries
selected_countries = ['Brazil', 'United States of America', 'Indonesia', 'Myanmar', 'India', 'China']
```

```python
# Melt the DataFrame to long format
ifl_melted = ifl.drop(columns=['iso']).melt(id_vars='country', var_name='Year', value_name='IFL')

# Clean Year column to strip the "IFL.raw." prefix
ifl_melted['Year'] = ifl_melted['Year'].str.strip('IFL.raw.')

# Filter the DataFrame to include only the selected countries
ifl_melted = ifl_melted[ifl_melted['country'].isin(selected_countries)]

# Create the plot
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='IFL', hue='country', data=ifl_melted, marker='o', linestyle='--', palette='pastel')
plt.title('Intact Forest Loss Over Time')
plt.ylabel('IFL')
plt.xlabel('Year')
plt.legend(title="Country")
plt.xticks(rotation=30)
plt.grid(True)
plt.show()
```

```python
pfl.head()
```

```python
pfl_melted = pfl.drop(columns=['iso']).melt(id_vars='country', var_name='Year', value_name='PFL')
pfl_melted['Year'] = pfl_melted['Year'].str.strip('PFL.raw.')
pfl_melted = pfl_melted[pfl_melted['country'].isin(selected_countries)]

# Create the plot
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='PFL', hue='country', data=pfl_melted,  marker='o', linestyle='--', palette='pastel')
plt.title('Primary Forest Loss Over Time')
plt.ylabel('PFL')
plt.xlabel('Year')
plt.legend(title="Country")
plt.xticks(rotation=30)
plt.grid(True)
plt.show()
```

```python
urb.head()
```

```python
urb_melted = urb.drop(columns=['iso']).melt(id_vars='country', var_name='Year', value_name='URB')
urb_melted['Year'] = urb_melted['Year'].str.strip('URB.raw.')
urb_melted = urb_melted[urb_melted['Year'] != '2005']
urb_melted = urb_melted[urb_melted['country'].isin(selected_countries)]

plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='URB', hue='country', data=urb_melted,  marker='o', linestyle='--', palette='pastel')
plt.title('Urban Population Percental (URB) Over Time')
plt.ylabel('Urban Population Percental (URB)')
plt.xlabel('Year')
plt.legend(title="Country")
plt.xticks(rotation=30)
plt.grid(True)
plt.show()
```
