```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.api import ExponentialSmoothing

# Load datasets
pfl = pd.read_csv("clean_PFL.csv")
ifl = pd.read_csv("clean_IFL.csv")
urb = pd.read_csv("clean_URB.csv")

# Reshape datasets from wide format to long format
def reshape_data(df, var_name):
    df_long = df.melt(id_vars=["country", "iso"], var_name="year", value_name=var_name)
    df_long["year"] = df_long["year"].str.extract("(\\d{4})").astype(int)  # Extract year from column names
    return df_long

pfl_long = reshape_data(pfl, "PFL.raw")
ifl_long = reshape_data(ifl, "IFL.raw")
urb_long = reshape_data(urb, "URB.raw")

# Merge datasets
data = pfl_long.merge(ifl_long, on=["country", "iso", "year"]).merge(urb_long, on=["country", "iso", "year"])

# Filter to common years (2006-2022)
data = data[(data["year"] >= 2006) & (data["year"] <= 2022)]

# Handle missing values
data.fillna(method='ffill', inplace=True)
data.dropna(inplace=True)

# Ensure only numeric columns are included for aggregation
numeric_cols = ["PFL.raw", "IFL.raw", "URB.raw"]

# Convert year column to datetime format
data["year"] = pd.to_datetime(data["year"], format='%Y')

# Aggregate data by year
yearly_data = data.groupby("year")[numeric_cols].mean()

# Cross-Correlation Analysis
def cross_corr_analysis(series1, series2, max_lag=5):
    """Computes cross-correlation between two time series with proper alignment."""
    lags = np.arange(-max_lag, max_lag + 1)
    correlations = [series1.corr(series2.shift(lag)) for lag in lags]

    # Plot
    plt.figure(figsize=(10, 5))
    plt.bar(lags, correlations, color="blue", alpha=0.7)
    plt.axhline(0, color="black", linewidth=1)
    plt.xlabel("Lag (years)")
    plt.ylabel("Cross-Correlation")
    plt.title(f"Cross-Correlation between {series1.name} and {series2.name}")
    plt.show()

    return lags[np.argmax(np.abs(correlations))], max(correlations, key=abs)

# Perform cross-correlation with aligned data
urb_series = yearly_data["URB.raw"]
pfl_series = yearly_data["PFL.raw"]
ifl_series = yearly_data["IFL.raw"]

best_lag_pfl, max_corr_pfl = cross_corr_analysis(urb_series, pfl_series)
best_lag_ifl, max_corr_ifl = cross_corr_analysis(urb_series, ifl_series)

print(f"Best lag for URB-PFL: {best_lag_pfl} years with correlation {max_corr_pfl:.3f}")
print(f"Best lag for URB-IFL: {best_lag_ifl} years with correlation {max_corr_ifl:.3f}")

# Forecasting Future Forest Loss using Exponential Smoothing
def forecast_exponential_smoothing(series, country, var_name, forecast_years=10):
    """Applies Holt's Linear Trend Exponential Smoothing to forecast future values for a given country."""
    model = ExponentialSmoothing(series, trend="add", seasonal=None, damped_trend=True)
    fit = model.fit()
    forecast = fit.forecast(forecast_years)
    r2 = 1 - (np.sum((series - fit.fittedvalues) ** 2) / np.sum((series - np.mean(series)) ** 2))  # Compute R-squared

    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(series.index, series, label="Observed", marker="o")
    plt.plot(pd.date_range(start=series.index[-1], periods=forecast_years+1, freq='Y')[1:],
             forecast, label="Forecast", linestyle="dashed", color="red")
    plt.xlabel("Year")
    plt.ylabel(var_name)
    plt.title(f"Forecasting {var_name} for {country} (R = {r2:.3f})")
    plt.legend()
    plt.show()

    return forecast, r2

# Define key countries for forecasting
key_countries = ["Brazil", "China", "India", "Indonesia", "Myanmar", "United States of America"]

# Perform forecasting for each country
forecast_results = {}
for country in key_countries:
    country_data = data[data["country"] == country].set_index("year")
    if not country_data.empty:
        forecast_results[country] = {}
        for var in ["PFL.raw", "IFL.raw"]:
            if var in country_data.columns:
                forecast, r2 = forecast_exponential_smoothing(country_data[var], country, var)
                forecast_results[country][var] = (forecast, r2)

# Identify Countries with Greatest Changes in Deforestation
def calculate_deforestation_change(data, var_name):
    """Calculates the total and relative change in deforestation over time for each country."""
    first_year = data["year"].min()
    last_year = data["year"].max()
    change_df = data.groupby("country").agg({var_name: ["first", "last"]})
    change_df.columns = ["start", "end"]
    change_df["absolute_change"] = change_df["end"] - change_df["start"]
    change_df["relative_change"] = (change_df["absolute_change"] / change_df["start"]) * 100
    return change_df.sort_values("absolute_change", ascending=False)

pfl_changes = calculate_deforestation_change(data, "PFL.raw")
ifl_changes = calculate_deforestation_change(data, "IFL.raw")

print("Countries with Greatest Increase in Primary Forest Loss:")
print(pfl_changes.head(10))
print("\nCountries with Greatest Increase in Intact Forest Loss:")
print(ifl_changes.head(10))

# Identify Countries at Greatest Risk for Future Deforestation
def identify_high_risk_countries(forecast_results):
    """Identifies countries predicted to have the highest future deforestation rates."""
    risk_scores = {}
    for country, forecasts in forecast_results.items():
        total_forecasted_loss = sum(forecasts["PFL.raw"][0]) + sum(forecasts["IFL.raw"][0])  # Sum of predicted loss
        risk_scores[country] = total_forecasted_loss
    return sorted(risk_scores.items(), key=lambda x: x[1], reverse=True)

high_risk_countries = identify_high_risk_countries(forecast_results)
print("\nCountries at Greatest Risk for Future Deforestation:")
print(high_risk_countries[:10])
```
